```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pywt  # ŸÖŸÉÿ™ÿ®ÿ© ŸÑÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸÖŸàŸäÿ¨ÿßÿ™
from scipy.fftpack import fft
from scipy.signal import butter, filtfilt, resample
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.metrics import f1_score, accuracy_score
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import validation_curve
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report



# üìÇ ÿ™ÿ≠ÿØŸäÿØ ŸÖÿ≥ÿßÿ± ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
data_path = r"D:\ÿπÿ¥Ÿàÿßÿ¶Ÿäÿßÿ™ ŸÑÿπŸÑÿß\WESAD\WESAD\PKL"

# üìå ÿ™ÿπÿ±ŸäŸÅ ÿßŸÑŸÖÿ¥ÿßÿ±ŸÉŸäŸÜ ŸÖŸÜ S2 ÿ•ŸÑŸâ S17
participant_ids = list(range(2, 18))

def remove_outliers_iqr(data):
    """ ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÖÿ™ÿ∑ÿ±ŸÅÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ IQR """
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data >= lower_bound) & (data <= upper_bound)]

def normalize_signal(signal):
    """ ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ•ÿ¥ÿßÿ±ÿ© ÿ®ŸäŸÜ 0 Ÿà 1 ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ Min-Max Scaling """
    return (signal - np.min(signal)) / (np.max(signal) - np.min(signal))
    
# üì• ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ WESAD ŸÑŸÉŸÑ ŸÖÿ¥ÿßÿ±ŸÉ
def load_wesad_data(participant_ids, data_path):
    data = {}
    for pid in participant_ids:
        file_path = os.path.join(data_path, f'S{pid}.pkl')
        if os.path.exists(file_path):
            try:
                with open(file_path, 'rb') as f:
                    raw_data = pickle.load(f, encoding='latin1')
                
                labels = raw_data.get('label', [])  
                ecg_signal = raw_data['signal']['chest']['ECG'].flatten()
                
                # ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ™ÿµŸÜŸäŸÅÿßÿ™
                labels_binary = np.where(labels == 2, 1, np.where(np.isin(labels, [1, 3, 4]), 0, np.nan))
                
                # ÿ•ÿ≤ÿßŸÑÿ© ÿßŸÑŸÇŸäŸÖ ÿ∫Ÿäÿ± ÿßŸÑŸÖÿ±ÿ∫Ÿàÿ®ÿ© ( 5ÿå 6ÿå 7)
                valid_indices = ~np.isnan(labels_binary) 
                labels_binary = labels_binary[valid_indices]
                ecg_signal = ecg_signal[valid_indices]
                
                # ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÖÿ≥ŸÖŸàÿ≠ ÿ®Ÿáÿß ÿ®ÿπÿØ IQR
                valid_ecg = remove_outliers_iqr(ecg_signal)
                iqr_mask = np.isin(ecg_signal, valid_ecg)
                # ÿßŸÑÿßÿ≠ÿ™ŸÅÿßÿ∏ ŸÅŸÇÿ∑ ÿ®ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÖÿ™ÿ≤ÿßŸÖŸÜÿ©
                ecg_signal = ecg_signal[iqr_mask]
                labels_binary = labels_binary[iqr_mask]

                # üîπ ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ•ÿ¥ÿßÿ±ÿ© ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ StandardScaler
                ecg_signal = normalize_signal(ecg_signal)
                
                data[pid] = {'ecg': ecg_signal, 'labels': labels_binary}
            except Exception as e:
                print(f"‚ö† ÿÆÿ∑ÿ£ ŸÅŸä ÿ™ÿ≠ŸÖŸäŸÑ {file_path}: {e}")
    return data

# ‚úÖ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
wesad_data = load_wesad_data(participant_ids, data_path)

# üîç ÿßŸÑÿ™ÿ≠ŸÇŸÇ ŸÖŸÜ ÿßŸÑŸÖÿ¥ÿßÿ±ŸÉŸäŸÜ ÿßŸÑÿ∞ŸäŸÜ ÿ™ŸÖÿ™ ŸÖÿπÿßŸÑÿ¨ÿ™ŸáŸÖ
print("üìä ÿßŸÑŸÖÿ¥ÿßÿ±ŸÉŸàŸÜ ÿßŸÑÿ∞ŸäŸÜ ÿ™ŸÖ ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ŸáŸÖ:", list(wesad_data.keys()))



# üîÑ ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ•ÿ¥ÿßÿ±ÿ© ÿ•ŸÑŸâ ŸÜŸàÿßŸÅÿ∞ ÿ≤ŸÖŸÜŸäÿ©
def segment_signal(signal, labels, window_size=10, sampling_rate=700):
    window_length = window_size * sampling_rate  
    num_windows = len(signal) // window_length   
    segments, segment_labels = [], []

    for i in range(num_windows):
        start_idx = i * window_length
        end_idx = start_idx + window_length
        segment = signal[start_idx:end_idx]
        segment_label = labels[start_idx:end_idx]  
        final_label = final_label = 1 if np.sum(segment_label) > len(segment_label) / 2 else 0 
        segments.append(segment)
        segment_labels.append(final_label)
    
    return np.array(segments), np.array(segment_labels)
    

# üîÑ ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ŸÑŸÉŸÑ ŸÖŸÇÿ∑ÿπ
def extract_features(segments):
    features = []
    for segment in segments:
        mean = np.mean(segment)
        std = np.std(segment)
        max_val = np.max(segment)
        min_val = np.min(segment)
        features.append([mean, std, max_val, min_val])
    return np.array(features)


    

# üöÄ ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑŸÅŸÑÿ™ÿ±ÿ©ÿå ÿßŸÑÿ™ŸÇÿ≥ŸäŸÖÿå Ÿàÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ŸÑŸÉŸÑ ŸÖÿ¥ÿßÿ±ŸÉ
processed_data = {}

for pid in wesad_data.keys():
    ecg_signal = bandpass_filter(wesad_data[pid]['ecg'])
    segments, segment_labels = segment_signal(ecg_signal, wesad_data[pid]['labels'])
    features = extract_features(segments)  # ‚úÖ ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿßŸÑŸÖÿ™ŸÇÿØŸÖÿ©

    processed_data[pid] = {
        'features': features,
        'labels': segment_labels
    }

# üî¢ ÿ≠ÿ≥ÿßÿ® ŸÜÿ≥ÿ®ÿ© ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖÿµŸÜŸÅÿ© ŸÉŸÄ Stressed ÿ•ŸÑŸâ Non-Stressed
stress_ratios = {}

for pid, pdata in processed_data.items():
    stress_count = np.sum(pdata['labels'] == 1)  # ÿπÿØÿØ ŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ∂ÿ∫ÿ∑
    non_stress_count = np.sum(pdata['labels'] == 0)  # ÿπÿØÿØ ŸÜŸàÿßŸÅÿ∞ ÿ∫Ÿäÿ± ÿßŸÑÿ∂ÿ∫ÿ∑
    stress_ratio = stress_count / (non_stress_count + 1)  # ÿ™ÿ¨ŸÜÿ® ÿßŸÑŸÇÿ≥ŸÖÿ© ÿπŸÑŸâ ÿµŸÅÿ±

    stress_ratios[pid] = stress_ratio

# üìä ÿπÿ±ÿ∂ ÿßŸÑŸÜÿ™ÿßÿ¶ÿ¨
for pid, ratio in stress_ratios.items():
    print(f"üìå ÿßŸÑŸÖÿ¥ÿßÿ±ŸÉ S{pid}: ŸÜÿ≥ÿ®ÿ© ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖÿ∂ÿ∫Ÿàÿ∑ÿ© ÿ•ŸÑŸâ ÿ∫Ÿäÿ± ÿßŸÑŸÖÿ∂ÿ∫Ÿàÿ∑ÿ© = {ratio:.2f}")




# üîπ ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖÿßÿ∞ÿ¨
models = {
    'KNN': KNeighborsClassifier(n_neighbors=50, weights='distance'),
    'SVM': SVC(kernel='rbf', C=15, gamma='scale', probability=True),
    'RF': RandomForestClassifier(n_estimators=100, min_samples_split=2, max_depth=10, random_state=42)
}

# üîÑ ÿ•ÿπÿØÿßÿØ LOPO-CV
logo = LeaveOneGroupOut()
groups = []  
X, y = [], []  
confusion_matrices = {}  # ÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ£ŸÜ ŸÖÿµŸÅŸàŸÅÿßÿ™ ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥ ŸÖÿ≠ÿØÿ´ÿ©

# ‚úÖ ÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÑÿ¨ŸÖŸäÿπ ÿßŸÑŸÖÿ¥ÿßÿ±ŸÉŸäŸÜ
for pid in processed_data.keys():
    X.extend(processed_data[pid]['features'])
    y.extend(processed_data[pid]['labels'])
    groups.extend([pid] * len(processed_data[pid]['labels']))

X, y = np.array(X), np.array(y)

# üöÄ ÿ™ÿ∑ÿ®ŸäŸÇ LOPO-CV ŸÑŸÉŸÑ ŸÜŸÖŸàÿ∞ÿ¨
results = {}
fpr_dict = {}
tpr_dict = {}
auc_dict = {}

for model_name, model in models.items():
    f1_scores = []
    accuracy_scores = []
    
    all_y_true = []
    all_y_pred = []
    
    # ÿ™ÿÆÿ≤ŸäŸÜ ŸÖÿµŸÅŸàŸÅÿßÿ™ ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥ ŸÑŸÉŸÑ ŸÜŸÖŸàÿ∞ÿ¨
    confusion_matrices[model_name] = []

    for train_idx, test_idx in logo.split(X, y, groups):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        # ‚öôÔ∏è ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
        scaler = StandardScaler()
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)
        
        # üü¢ 1Ô∏è‚É£ ÿ≤ŸäÿßÿØÿ© ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™Ÿàÿ™ÿ± ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ SMOTE
        smote = SMOTE(sampling_strategy='auto', random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
        
        # ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
        model.fit(X_train_resampled, y_train_resampled)
        
        # ÿßŸÑÿ™ŸÜÿ®ÿ§
        y_pred = model.predict(X_test)
        
        # ‚úÖ ÿ™ÿÆÿ≤ŸäŸÜ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÅÿπŸÑŸäÿ© ŸàÿßŸÑŸÖÿ™ŸàŸÇÿπÿ©
        all_y_true.extend(y_test)
        all_y_pred.extend(y_pred)
        
        # ÿ≠ÿ≥ÿßÿ® ÿßŸÑÿØŸÇÿ© Ÿà F1-score
        acc = accuracy_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred, average='macro')
        
        accuracy_scores.append(acc)
        f1_scores.append(f1)

        # ÿ≠ÿ≥ÿßÿ® FPR Ÿà TPR Ÿà AUC
        fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:, 1])
        roc_auc = auc(fpr, tpr)
        
        # ÿ™ÿÆÿ≤ŸäŸÜ ÿ®ŸäÿßŸÜÿßÿ™ ROC ŸÑŸÉŸÑ ŸÜŸÖŸàÿ∞ÿ¨
        fpr_dict[model_name] = fpr
        tpr_dict[model_name] = tpr
        auc_dict[model_name] = roc_auc

        # ÿ≠ÿ≥ÿßÿ® ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
        cm = confusion_matrix(y_test, y_pred)
        confusion_matrices[model_name].append(cm)

    # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÖÿ™Ÿàÿ≥ÿ∑ ŸÑŸÑŸÜŸÖŸàÿ∞ÿ¨
    results[model_name] = {
        'accuracy': np.mean(accuracy_scores),
        'f1_score': np.mean(f1_scores)
    }
    
    print(f"\nüìÑ ÿßŸÑÿ™ÿµŸÜŸäŸÅ ÿßŸÑŸÜŸáÿßÿ¶Ÿä ŸÑŸÜŸÖŸàÿ∞ÿ¨ {model_name}:")
    print(classification_report(all_y_true, all_y_pred, digits=4))

    
    print(f"‚úÖ {model_name} - Accuracy: {results[model_name]['accuracy']:.2f}, F1-score: {results[model_name]['f1_score']:.2f}")

    # üîç ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿπÿØÿØ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÅÿπŸÑŸäÿ© ŸàÿßŸÑŸÖÿ™ŸàŸÇÿπÿ©
    print(f"üîç ÿπÿØÿØ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÅÿπŸÑŸäÿ©: {len(all_y_true)}, ÿπÿØÿØ ÿßŸÑŸÇŸäŸÖ ÿßŸÑŸÖÿ™ŸàŸÇÿπÿ©: {len(all_y_pred)}")

# üé® ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸâ ROC ŸÑŸÉŸÑ ŸÜŸÖŸàÿ∞ÿ¨
plt.figure(figsize=(10, 8))
for model_name in models.keys():
    plt.plot(fpr_dict[model_name], tpr_dict[model_name], label=f'{model_name} (AUC = {auc_dict[model_name]:.2f})')

# ÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿ∑ ÿßŸÑÿπÿ¥Ÿàÿßÿ¶Ÿä
plt.plot([0, 1], [0, 1], 'k--', label='Random')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

# üî• ÿ∑ÿ®ÿßÿπÿ© ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥ ŸÑŸÉŸÑ ŸÜŸÖŸàÿ∞ÿ¨
for model_name, cm_list in confusion_matrices.items():
    cm = np.sum(cm_list, axis=0)  # ÿ¨ŸÖÿπ ŸÉŸÑ ŸÖÿµŸÅŸàŸÅÿßÿ™ ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}")
    plt.show()
```
